```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
torch.manual_seed(1234)

# Note: The default behavior now has injection attack prevention off.
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen-VL-Chat-Int4", trust_remote_code=True)

# use cuda device
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen-VL-Chat-Int4", device_map="cuda", trust_remote_code=True).eval()
```

```python
img='https://drive.google.com/uc?export=view&id=19lBhhvAvoasL-zVwBRK9WqaG1cIjAcsM'
```

```python
# 1st dialogue turn
query = tokenizer.from_list_format([
    {'image': img},
    {'text': 'What specific symbol is present in this tattoo? '},
])
response, history = model.chat(tokenizer, query=query, history=None)
print(response)
```
